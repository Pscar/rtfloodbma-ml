```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
from pandas import json_normalize

```

```python
url='https://www.rtfloodbma.com/api/rains';
data_rains=pd.read_json(url,orient='columns')
data_rains
```

```python
url='https://www.rtfloodbma.com/api/waters';
data_waters=pd.read_json(url,orient='columns')
data_waters
```

```python
station=pd.DataFrame(data_rains.data.values.tolist())['station']
pd.DataFrame(station).from_records(station)
pd.json_normalize(station)
```

```python
waters = pd.DataFrame(data_waters.data.values.tolist())
pd.DataFrame(waters).from_records(waters).head()
```

```python
rainfall_datetime = pd.DataFrame(data_rains.data.values.tolist())['rainfall_datetime']
rainfall_datetime
```

```python
rain_24h = pd.DataFrame(data_rains.data.values.tolist())['rain_24h']
rain_24h
```

```python
waterlevel_msl = pd.DataFrame(data_waters.data.values.tolist())['waterlevel_msl']
waterlevel_msl
```

```python
x=pd.concat([rain_24h, waterlevel_msl])
x
```

```python
dc=pd.concat([pd.Series(rainfall_datetime,name="date"),pd.DataFrame(json_normalize(station))],axis=1)
```

```python
table = pd.concat([dc,x],ignore_index=False) 
```

```python
table.drop(columns=['id','tele_station_oldcode','tele_station_type','tele_station_lat','tele_station_long','sub_basin_id'],axis=1)
```

```python
table.drop(columns=['id','tele_station_oldcode','tele_station_type','tele_station_lat','tele_station_long','sub_basin_id'],axis=1).to_csv('out4.csv')
```
